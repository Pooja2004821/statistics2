Q1.What is hypothesis testing in statistics?
- Hypothesis testing in statistics is a formal procedure used to make inferences or decisions about a population based on sample data. It helps determine whether there is enough evidence to support a specific claim or hypothesis about a population parameter.
- Key Concepts of Hypothesis Testing
  - Null Hypothesis (H₀) : This is the default assumption or claim.
      - Example: “The average height of students is 160 cm.”
  - Alternative Hypothesis (H₁ or Ha) : This is what you want to prove or the opposing claim.
      - Example: “The average height of students is not 160 cm.”
  - Test Statistic : A value calculated from sample data used to test the hypothesis (e.g., Z, t, chi-square).
  - Significance Level (α) : The probability of rejecting the null hypothesis when it is actually true.Common values: 0.05, 0.01.
  - P-value : The probability of observing the test statistic or something more extreme under the null hypothesis.If p-value < α, reject the null hypothesis.
  - Decision : Reject H₀ if the p-value is less than α (significant result).Fail to reject H₀ if the p-value is greater than α (not enough evidence).

Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis is the default or starting assumption in hypothesis testing. It states that there is no effect, no difference, or no relationship between variables.It represents the status quo or a claim to be tested.The burden of proof is on disproving (rejecting) it.
- Example : If a company claims that a new drug has no effect on blood pressure, the null hypothesis would be:
H₀: μ = μ₀ → The mean blood pressure is the same as before (no change).
- The alternative hypothesis is the statement you want to test or prove. It contradicts the null hypothesis and suggests that there is an effect, a difference, or a relationship.It's accepted if there is sufficient evidence to reject H₀.
- Example (continued) : The alternative hypothesis might be:
H₁: μ ≠ μ₀ → The mean blood pressure is different (drug has an effect).
- Key Differences Between H₀ and H₁
Feature................Null Hypothesis (H₀)..........Alternative Hypothesis (H₁)
Meaning.................No change or effect..........Some change or effect
Role..........Assumed true unless evidence says otherwise...Tested against
Symbol........................H₀.........................H₁ or Ha
Type of test..............Equality (e.g., =, ≤, ≥).........Inequality (≠, <, >)
Goal in testing.............Try to reject it..............Try to support it

Q3.What is the significance level in hypothesis testing, and why is it important?
- What is the Significance Level (α) in Hypothesis Testing?
The significance level, denoted by α (alpha), is the threshold probability used to decide whether to reject the null hypothesis (H₀).It represents the risk you're willing to take of making a Type I error—that is, rejecting the null hypothesis when it's actually true.
- Common Values of α:
α(Significance Level)..................Confidence Level............Meaning
0.05....................95%........ 5% risk of a false positive (Type I error)
0.01................99%...............................More strict, only 1% risk
0.10...............................90%.................Less strict, 10% risk
- Why Is Significance Level Important?
  - Controls Type I Error Rate : α sets how much error you're willing to tolerate by rejecting a true H₀.
    - Example: α = 0.05 means you’ll wrongly reject H₀ in 5 out of 100 tests (on average) even if it's actually true.
  - Decision Making : It serves as a benchmark to compare the p-value against.If p-value < α, you reject H₀.If p-value ≥ α, you fail to reject H₀.
  - Affects Test Sensitivity : Lower α → Stricter test, fewer false positives, but more false negatives (Type II errors).Higher α → More lenient test, easier to detect effect, but higher risk of false positives.

Q4. What does a P-value represent in hypothesis testing?
- The P-value (probability value) is the probability of obtaining results at least as extreme as the observed data, assuming that the null hypothesis (H₀) is true.In simple words, it tells you how surprising or unlikely your result is under the assumption that H₀ is true.
- Interpretation of P-value : Small p-value (≤ α): Strong evidence against H₀ → Reject H₀.Large p-value (> α): Weak evidence against H₀ → Fail to reject H₀

Q5. How do you interpret the P-value in hypothesis testing?
- The p-value tells you how likely it is to get your sample data—or something more extreme—if the null hypothesis (H₀) is actually true.
- Interpretation Rule :
  - If p-value ≤ α (significance level) → Reject H₀
  - The result is statistically significant.
  - There is enough evidence to support the alternative hypothesis (H₁).
  - If p-value > α → Fail to reject H₀
  - The result is not statistically significant.
  - There is not enough evidence to support H₁.

Q6. What are Type 1 and Type 2 errors in hypothesis testing?
- When you perform a hypothesis test, you’re making a decision based on sample data — and like all decisions, there's a chance of making a mistake. These mistakes are called Type I and Type II errors.
- Type I Error (False Positive) :
  - You reject the null hypothesis (H₀) when it is actually true.
  - You think there’s an effect when in reality, there isn’t.
  - The probability of making a Type I error is α (significance level).
  - Example :
       - A medical test says a healthy person is sick.
       - H₀: Person is healthy
       - You reject H₀ → Mistake!
- Type II Error (False Negative) :
  - You fail to reject the null hypothesis when it is actually false.
  - You miss a real effect that actually exists.
  - The probability of a Type II error is β.
  - The power of a test is 1 − β (i.e., the ability to detect an effect when it exists).
  - Example:
      - A medical test fails to detect illness in a sick person.
      - H₀: Person is healthy
      - You do not reject H₀ → Mistake!

Q7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- In hypothesis testing, whether a test is one-tailed or two-tailed depends on the direction of the claim you're testing in the alternative hypothesis (H₁).
1. One-Tailed Test : A one-tailed test checks for an effect in only one direction—either greater than or less than a certain value.
- Used when : You are only interested in one side of the distribution.
- Example :
         - H₀: μ = 100
         - H₁: μ > 100 (right-tailed) → you're testing if the mean is greater
         - H₁: μ < 100 (left-tailed) → you're testing if the mean is less
- Graph : The entire α (significance level) is in one tail of the distribution.
2. Two-Tailed Test : A two-tailed test checks for an effect in both directions—either greater than or less than a certain value.
Used when : You're interested in any difference, regardless of direction.
- Example :
         - H₀: μ = 100
         - H₁: μ ≠ 100 → you're testing if the mean is different, whether higher or lower
- Graph : The α is split into two tails (α/2 in each tail).

Q8.What is the Z-test, and when is it used in hypothesis testing?
- A Z-test is a statistical test used to determine whether there is a significant difference between sample and population means (or between two samples), assuming the population standard deviation is known.It is based on the Z-distribution (standard normal distribution), which has a mean of 0 and standard deviation of 1.
- When to Use a Z-test
    - You use a Z-test when:
    - The sample size is large (typically n ≥ 30)
    - The population standard deviation σ is known
    - The data is approximately normally distributed (or the sample size is large enough for CLT to apply)

Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- The Z-score is a way to standardize your test statistic and tell how far (in standard deviations) a data point or sample mean is from the population mean, under the null hypothesis.
- What the Z-score Represents : The Z-score tells you how many standard errors the sample mean is away from the population mean.
     - Z = 0 → Sample mean is equal to the population mean
     - Z > 0 → Sample mean is above the population mean
     - Z < 0 → Sample mean is below the population mean
- In Hypothesis Testing : You compare the Z-score to a critical value (from Z-tables) based on the significance level (α)Or, you use the Z-score to find the p-value
- Decision Rule (for α = 0.05 in a two-tailed test) :
     - If |Z| > 1.96 → Reject H₀
     - If |Z| ≤ 1.96 → Fail to reject H₀

Q10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution, also called Student’s t-distribution, is a probability distribution that is similar to the normal distribution but has heavier tails. It’s used when estimating population parameters when the sample size is small and/or the population standard deviation (σ) is unknown.
- Key Characteristics of the T-distribution :  
   - Symmetrical and bell-shaped, like the normal distribution.
   - Flatter and wider than the normal distribution → more area in the tails.
   - As the sample size increases, the T-distribution approaches the normal distribution.
   - Depends on degrees of freedom (df) = n−1
- When to Use the T-distribution (Instead of Normal/Z-distribution) :
   - Condition.............................................Use T-distribution?
   - Population standard deviation unknown.................Yes
   - Sample size is small (n < 30).........................Yes
   - Sample drawn from approximately normal population.....Yes
   - Large sample size and known σ.........................Use Z-distribution
- Formula for T-score (One-sample T-test) :
           - t= (X bar − μ)/(s/(root n))
           - Where:
                  - X bar = sample mean
                  - μ = population mean (under H₀)
                  - s = sample standard deviation
                  - n = sample size
                  - Degrees of freedom = n−1

Q13. What is the relationship between Z-test and T-test in hypothesis testing?
- Here’s the relationship between Z-test and T-test in hypothesis testing, in simple points :
   - Both are used to test hypotheses about means—whether a sample mean differs from a population mean or between two sample means.
   - Both calculate a test statistic (Z or t) that measures how far the sample statistic is from the hypothesized population parameter, in standardized units.
   - The Z-test uses the normal distribution, assuming the population standard deviation (σ) is known and the sample size is large.
   - The T-test uses the t-distribution, which adjusts for extra uncertainty when the population standard deviation is unknown and the sample size is small.
   - As the sample size increases, the t-distribution approaches the normal distribution, so the T-test statistic behaves more like the Z-test statistic.
   - Essentially, the T-test generalizes the Z-test to situations where σ is unknown and small samples are involved.

Q14.What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (like a mean or proportion) with a certain level of confidence.
- Key Points :
   - The confidence level (e.g., 95%) represents how sure we are that the interval contains the true population parameter.
   - A 95% confidence interval means: If we took many samples and built intervals the same way, about 95% of those intervals would contain the true parameter.
   - The interval is expressed as : Estimate±Margin of Error
   - The margin of error depends on the standard error and the critical value from the Z or t-distribution.
- How is it Used to Interpret Statistical Results?
   - Provides a range estimate: Instead of just a point estimate (e.g., sample mean = 50), CI gives a range (e.g., 47 to 53) where the true mean likely lies.
   - Shows precision : Narrower intervals mean more precise estimates; wider intervals mean less precision.
   - Helps in hypothesis testing : If the confidence interval does not contain the hypothesized value (e.g., population mean), it suggests significant difference and you may reject H₀.If it contains the hypothesized value, you fail to reject H₀.

Q15. What is the margin of error, and how does it affect the confidence interval?
- The margin of error (ME) is the amount added to and subtracted from a sample estimate to create a confidence interval. It represents the maximum expected difference between the true population parameter and the sample estimate due to sampling variability.
- How to Calculate Margin of Error :
        - Margin of Error=Critical value×Standard error
             - Where :
                 - Critical value comes from the Z-distribution or t-distribution (depends on confidence level and sample size)
                 - Standard error (SE) is the standard deviation of the sampling distribution, e.g., for a mean :
                       - SE = s/(root n) (s is sample standard deviation, n is sample size)
- How Does Margin of Error Affect the Confidence Interval?
   - The confidence interval is calculated as  :
            - Estimate±Margin of Error
   - A larger margin of error → wider confidence interval → less precise estimate
   - A smaller margin of error → narrower confidence interval → more precise estimate
- Factors Affecting Margin of Error:
   - Confidence Level: Higher confidence (e.g., 99% vs 95%) → larger critical value → larger margin of error
   - Sample Size: Larger sample size → smaller standard error → smaller margin of error
   - Variability in Data: More variability (larger standard deviation) → larger standard error → larger margin of error

     - Where:

         - P(B) = Total probability of evidence B
- How is it used in statistics?

  - Combines prior knowledge with new evidence to make better probabilistic predictions.

Q17. What is the Chi-square distribution, and when is it used?

- Key Characteristics :
  - Always non-negative (χ² ≥ 0)
  - Shape depends on the degrees of freedom (df)
  - Skewed right for small df; becomes more normal as df increases
  - Only defined for positive values
- When is the Chi-square Distribution Used?

- Common Chi-square Tests :
1. Chi-square Goodness-of-Fit Test
- Use : To test if observed frequencies match expected frequencies for one categorical variable.
- Example : Is a die fair?
2. Chi-square Test of Independence
- Use : To test if two categorical variables are independent.
- Example : Is gender independent of voting preference?
3. Chi-square Test for Homogeneity
Use : To compare distributions across different populations.
Example : Do three schools have the same distribution of favorite subjects?

Q20. What is an ANOVA test, and what are its assumptions?
- ANOVA (Analysis of Variance) is a statistical method used to determine whether there are significant differences between the means of three or more independent groups.It compares the variance between groups to the variance within groups to test the null hypothesis that all group means are equal.
- Hypotheses in ANOVA :
   - Null hypothesis (H₀): All group means are equal 𝜇1=𝜇2=𝜇3=⋯=𝜇𝑘
   - Alternative hypothesis (H₁): At least one mean is different
   - F-statistic Formula in One-way ANOVA :
          - F = (Variance within groups)/(Variance between groups) = (MSW)(MSB)
                  - MSB: Mean Square Between groups
                  - MSW: Mean Square Within groups
- When to Use ANOVA : You have one categorical independent variable with 3 or more groups.You are testing its effect on a continuous dependent variable
- Example: Comparing average exam scores across 4 different teaching methods.
- Assumptions of ANOVA :
- Assumption........................................................Explanation
1. Independence............Observations are independent across and within groups
2. Normality.....Data in each group should be approximately normally distributed
3. Homogeneity of variances.......All groups should have roughly equal variances

Q21. What are the different types of ANOVA tests?
- ANOVA (Analysis of Variance) tests come in several types, depending on the number of independent variables (factors) and whether repeated measures are involved.
1. One-Way ANOVA (Single-Factor ANOVA)
- Use when: You have one independent variable (factor) with 3 or more levels (groups).
- Purpose: Test if group means differ significantly.
- Example: Compare test scores from 3 different teaching methods.
               - 𝐻0:𝜇1=𝜇2=𝜇3

- Use when: You have two independent variables (factors).
- Purpose: Test for :
     - Main effect of Factor A
     - Main effect of Factor B
     - Interaction between A and B
- Example: Test how teaching method (A) and study time (B) affect test scores.

- Use when: The same subjects are measured multiple times (within-subjects design).

4. Mixed-Design ANOVA (Split-Plot ANOVA)
- Use when: You have both :
       - Between-subject factors (e.g., different groups)
       - Within-subject factors (e.g., repeated measures)
- Example: Different treatment groups (between-subject) are tested at multiple time points (within-subject).
5. MANOVA (Multivariate Analysis of Variance)
- Use when : You have multiple dependent variables.
- Purpose : Determine if groups differ across a combination of outcomes.
- Example : Comparing students across schools based on both math and science scores simultaneously.

```python
#PRACTICAL QUESTIONS
#Q1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from scipy import stats

# Sample data
sample = [102, 100, 98, 105, 97, 96, 104, 101, 99, 100]
sample_mean = np.mean(sample)
sample_size = len(sample)

# Known population parameters
population_mean = 100      # µ₀
population_std = 3         # σ (population standard deviation)

# Significance level
alpha = 0.05

# Z-test statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# P-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Output results
print(f"Sample Mean: {sample_mean}")
print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference between sample and population mean.")
```

```python
#Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(pop_mean, alt_mean, pop_std, sample_size, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors for a one-sample Z-test.

    Parameters:
    - pop_mean: Mean under null hypothesis (H0)
    - alt_mean: Mean under alternative hypothesis (H1)
    - pop_std: Known population standard deviation
    - sample_size: Sample size (n)
    - alpha: Significance level (default 0.05)
    """
    # Calculate critical z value for alpha (two-tailed test)
    z_crit = norm.ppf(1 - alpha/2)

    # Calculate critical sample mean boundaries under H0
    se = pop_std / np.sqrt(sample_size)
    lower_crit = pop_mean - z_crit * se
    upper_crit = pop_mean + z_crit * se

    # Define x range covering both distributions
    x_min = pop_mean - 4*se
    x_max = max(pop_mean, alt_mean) + 4*se
    x = np.linspace(x_min, x_max, 1000)

    # PDFs under H0 and H1
    pdf_h0 = norm.pdf(x, pop_mean, se)
    pdf_h1 = norm.pdf(x, alt_mean, se)

    # Calculate Type 2 error (beta): area under H1 curve between critical limits
    beta_region = (x > lower_crit) & (x < upper_crit)
    beta = np.trapz(pdf_h1[beta_region], x[beta_region])

    # Plot PDFs
    plt.figure(figsize=(10,6))
    plt.plot(x, pdf_h0, label='Null Hypothesis (H0)', color='blue')
    plt.plot(x, pdf_h1, label='Alternative Hypothesis (H1)', color='green')

    # Shade Type 1 error region (α): tails under H0 beyond critical values
    plt.fill_between(x, 0, pdf_h0, where=(x < lower_crit), color='red', alpha=0.3, label='Type 1 Error (α)')
    plt.fill_between(x, 0, pdf_h0, where=(x > upper_crit), color='red', alpha=0.3)

    # Shade Type 2 error region (β): area under H1 between critical values
    plt.fill_between(x, 0, pdf_h1, where=beta_region, color='orange', alpha=0.3, label='Type 2 Error (β)')

    # Critical value lines
    plt.axvline(lower_crit, color='black', linestyle='--', label=f'Critical Value Lower = {lower_crit:.2f}')
    plt.axvline(upper_crit, color='black', linestyle='--', label=f'Critical Value Upper = {upper_crit:.2f}')

    plt.title('Type 1 and Type 2 Errors Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Significance Level (Type 1 error, α): {alpha}")
    print(f"Calculated Type 2 error (β): {beta:.4f}")
    print(f"Power of test (1 - β): {1 - beta:.4f}")

# Example usage:
plot_type1_type2_errors(
    pop_mean=100,      # Null hypothesis mean
    alt_mean=105,      # Alternative hypothesis mean
    pop_std=10,        # Population standard deviation
    sample_size=30,    # Sample size
    alpha=0.05         # Significance level
)
```

```python
#Q9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy import stats

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - sample: list or array of sample data
    - confidence: confidence level (default is 0.95)

    Returns:
    - Tuple: (lower_bound, upper_bound)
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = stats.sem(sample)  # Standard error of the mean
    margin = stats.t.ppf((1 + confidence) / 2.0, n - 1) * std_err

    return mean - margin, mean + margin
# Simulated sample data
data = [70, 72, 68, 74, 71, 69, 73, 75, 70, 72]

ci_lower, ci_upper = confidence_interval(data, confidence=0.95)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Q10.Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy import stats

def margin_of_error(sample, confidence=0.95):
    """
    Calculate margin of error for the sample mean at the given confidence level.

    Parameters:
    - sample: list or array of sample values
    - confidence: confidence level (default = 0.95 for 95%)

    Returns:
    - margin of error (float)
    """
    n = len(sample)
    std_err = stats.sem(sample)  # Standard error of the mean
    t_score = stats.t.ppf((1 + confidence) / 2.0, df=n-1)  # t-score for confidence level
    moe = t_score * std_err
    return moe

# ---------------- Sample Usage ----------------
# Simulated sample data
data = [55, 60, 58, 62, 57, 59, 61, 56, 63, 60]

# Calculate margin of error
moe_95 = margin_of_error(data, confidence=0.95)
print(f"Margin of Error (95% confidence): {moe_95:.2f}")

moe_99 = margin_of_error(data, confidence=0.99)
print(f"Margin of Error (99% confidence): {moe_99:.2f}")
```

```python
#Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior, sensitivity, specificity):
    """
    Implements Bayes' Theorem for a binary classification.

    Parameters:
    - prior: P(H) - prior probability of the hypothesis (e.g. having disease)
    - sensitivity: P(D|H) - true positive rate
    - specificity: P(~D|~H) - true negative rate

    Returns:
    - posterior: P(H|D) - updated probability of hypothesis given positive result
    """
    # P(D|~H): False positive rate
    false_positive_rate = 1 - specificity

    # P(D): Total probability of positive test
    evidence = (sensitivity * prior) + (false_positive_rate * (1 - prior))

    # P(H|D): Posterior
    posterior = (sensitivity * prior) / evidence
    return posterior

# Inputs
prior = 0.001        # 0.1% prior probability (1 in 1000)
sensitivity = 0.99   # True positive rate
specificity = 0.95   # True negative rate

posterior = bayes_theorem(prior, sensitivity, specificity)
print(f"Probability of actually having the disease given a positive test: {posterior:.4f}")
```

```python
#Q16. Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    # Convert inputs to numpy arrays (handles lists too)
    sample1 = np.array(sample1)
    sample2 = np.array(sample2)

    # Check sample sizes
    if len(sample1) < 2 or len(sample2) < 2:
        raise ValueError("Each sample must have at least 2 observations.")

    # Calculate sample variances (unbiased estimator, ddof=1)
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Determine numerator and denominator for F-statistic
    if var1 >= var2:
        F = var1 / var2
        dfn = len(sample1) - 1
        dfd = len(sample2) - 1
    else:
        F = var2 / var1
        dfn = len(sample2) - 1
        dfd = len(sample1) - 1

    # Calculate two-tailed p-value
    p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))
    p_value = min(p_value, 1.0)  # p-value can't exceed 1

    # Interpret result
    if p_value < alpha:
        conclusion = "Reject null hypothesis: variances are significantly different."
    else:
        conclusion = "Fail to reject null hypothesis: no significant difference in variances."

    return F, p_value, conclusion

# Example usage
try:
    np.random.seed(42)
    sample1 = np.random.normal(100, 10, 30)
    sample2 = np.random.normal(100, 15, 30)

    F_stat, p_val, result = f_test(sample1, sample2)
    print(f"F-statistic: {F_stat:.4f}")
    print(f"P-value: {p_val:.4f}")
    print(f"Result: {result}")
except Exception as e:
    print("Error:", e)
```

```python
#Q19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups, alpha=0.05):
    """
    Check ANOVA assumptions for multiple groups:
    1. Normality within each group (Shapiro-Wilk test)
    2. Equal variances across groups (Levene's test)
    3. Independence - reminder (cannot be tested programmatically)

    Parameters:
    - groups: variable number of arrays/lists representing groups
    - alpha: significance level (default 0.05)

    Returns:
    - Dictionary with results of normality and variance tests
    """
    results = {
        "normality": {},
        "equal_variance": None,
        "independence_reminder": "Assume independence based on study design; not tested here."
    }

    # Normality test for each group
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        results["normality"][f"group_{i}"] = {
            "W_statistic": stat,
            "p_value": p,
            "normal": p > alpha
        }

    # Equal variance test (Levene’s test)
    stat, p = levene(*groups)
    results["equal_variance"] = {
        "Levene_statistic": stat,
        "p_value": p,
        "equal_variance": p > alpha
    }

    return results

# Example usage:
group1 = [23, 20, 22, 21, 19]
group2 = [30, 29, 31, 28, 32]
group3 = [25, 27, 26, 24, 28]

assumption_results = check_anova_assumptions(group1, group2, group3)

print("Normality Test Results:")
for group, res in assumption_results["normality"].items():
    print(f" {group}: p-value = {res['p_value']:.4f} -> {'Normal' if res['normal'] else 'Not Normal'}")

print("\nEqual Variance Test (Levene's Test):")
levene_res = assumption_results["equal_variance"]
print(f" p-value = {levene_res['p_value']:.4f} -> {'Equal variances' if levene_res['equal_variance'] else 'Unequal variances'}")

print("\nIndependence Reminder:")
print(assumption_results["independence_reminder"])
```

```python
#Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
from scipy.stats import ttest_ind

# Set random seed for reproducibility
np.random.seed(42)

# Simulate two samples from normal distributions
sample1 = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, SD=5, n=30
sample2 = np.random.normal(loc=52, scale=5, size=30)  # Mean=52, SD=5, n=30

# Perform two-sample independent t-test (testing if means differ)
t_stat, p_val = ttest_ind(sample1, sample2)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: The means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the means.")
```

```python
#Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
from scipy.stats import chi2

# Sample data (simulate or provide your own)
data = np.array([10.2, 9.8, 10.5, 10.0, 9.7, 10.3, 9.9, 10.1, 9.6, 10.4])

n = len(data)
sample_variance = np.var(data, ddof=1)  # Sample variance (unbiased)
sigma0_squared = 0.04  # Hypothesized population variance (for example)

# Calculate test statistic
chi_square_stat = (n - 1) * sample_variance / sigma0_squared

# Degrees of freedom
df = n - 1

# Significance level
alpha = 0.05

# Calculate critical values for two-tailed test
chi2_lower = chi2.ppf(alpha/2, df)
chi2_upper = chi2.ppf(1 - alpha/2, df)

# Calculate p-value
p_value = 2 * min(chi2.cdf(chi_square_stat, df), 1 - chi2.cdf(chi_square_stat, df))

print(f"Sample variance: {sample_variance:.4f}")
print(f"Chi-square statistic: {chi_square_stat:.4f}")
print(f"Critical values: lower = {chi2_lower:.4f}, upper = {chi2_upper:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision
if chi_square_stat < chi2_lower or chi_square_stat > chi2_upper:
    print("Reject the null hypothesis: The population variance is significantly different from hypothesized.")
else:
    print("Fail to reject the null hypothesis: Not enough evidence to say variance differs from hypothesized.")
```

```python
#Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
from statsmodels.stats.proportion import proportions_ztest

# Example data:
successes = [45, 30]  # Number of successes in group 1 and group 2
nobs = [100, 80]      # Number of observations (sample size) in group 1 and group 2

# Perform two-proportion z-test
stat, pval = proportions_ztest(successes, nobs)

print(f"Z-statistic: {stat:.4f}")
print(f"P-value: {pval:.4f}")

alpha = 0.05
if pval < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between proportions.")
```

```python
#Q26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Sample datasets (you can replace with your own data)
np.random.seed(0)
data1 = np.random.normal(loc=10, scale=2, size=30)  # Sample 1
data2 = np.random.normal(loc=10, scale=3, size=35)  # Sample 2

# Sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Sample sizes
n1 = len(data1)
n2 = len(data2)

# Calculate F statistic (ratio of larger variance over smaller variance)
if var1 > var2:
    F = var1 / var2
    dfn = n1 - 1  # degrees of freedom numerator
    dfd = n2 - 1  # degrees of freedom denominator
else:
    F = var2 / var1
    dfn = n2 - 1
    dfd = n1 - 1

# Significance level
alpha = 0.05

# Critical values for two-tailed test
F_critical_low = f.ppf(alpha / 2, dfn, dfd)
F_critical_high = f.ppf(1 - alpha / 2, dfn, dfd)

# P-value for two-tailed test
if F > 1:
    p_value = 2 * (1 - f.cdf(F, dfn, dfd))
else:
    p_value = 2 * f.cdf(F, dfn, dfd)

print(f"Variance 1: {var1:.4f}, Variance 2: {var2:.4f}")
print(f"F-statistic: {F:.4f}")
print(f"Degrees of freedom: numerator = {dfn}, denominator = {dfd}")
print(f"Critical F values: low = {F_critical_low:.4f}, high = {F_critical_high:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")

# Visualization of F-distribution and test statistic
x = np.linspace(0, 5, 500)
y = f.pdf(x, dfn, dfd)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label=f'F-distribution (dfn={dfn}, dfd={dfd})')

# Shade rejection regions
plt.fill_between(x, 0, y, where=(x <= F_critical_low), color='red', alpha=0.3, label='Rejection region (low)')
plt.fill_between(x, 0, y, where=(x >= F_critical_high), color='red', alpha=0.3, label='Rejection region (high)')

# Mark observed F-statistic
plt.axvline(F, color='black', linestyle='--', label=f'Observed F = {F:.4f}')

plt.title('F-Distribution with Critical Regions and Observed F Statistic')
plt.xlabel('F value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results
import numpy as np
from scipy.stats import chisquare

# Simulate observed counts for 4 categories
observed = np.array([50, 30, 10, 10])

# Define expected proportions (must sum to 1)
expected_proportions = np.array([0.4, 0.3, 0.2, 0.1])

# Calculate expected counts based on total observed count
total_count = observed.sum()
expected = expected_proportions * total_count

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Observed counts: {observed}")
print(f"Expected counts: {expected.astype(int)}")
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed data fits the expected distribution well.")
```
